```python
import math
import numpy as np

def prom_update(prom_prev: float, x_new: float, n_curr: int):
    if n_curr < 1:
        raise Exception()
    elif n_curr == 1:
        return x_new
    else:
        return prom_prev + (x_new - prom_prev) / n_curr


def s_cuad_update(s_cuad_prev: float, prom_prev: float, prom_curr: float, n_curr: int):
    if n_curr < 1:
        raise Exception()
    elif n_curr == 1:
        return 0.0
    else:
        return (1 - (1/(n_curr - 1))) * s_cuad_prev + n_curr * ((prom_curr - prom_prev)**2)
```

### Ejercicio 2.
Ejercicio 2: Estimar mediante el método de Monte Carlo la integral

1.
$$
\int_{0}^{1} \frac{e^x}{\sqrt{2x}} \ dx
$$

2.
$$
\int_{-\infty}^{\infty} x^2 \exp(-x^2) \ dx
$$

a) Generar al menos $100$ valores y detener la simulación cuando la desviación estándar muestral del estimador sea menor que $0.01$.


---

```python
def punto_i(u: float):
    return math.exp(u) / math.sqrt(2 * u)

def punto_ii(u: float):
    def g(x: float):
        return (x**2) * math.exp(-(x**2))

    return g(1/u - 1) / (u**2)

def sim_i():
    promedio = punto_i(np.random.random())
    s_cuad, n = 0, 1

    while n <= 100 or math.sqrt(s_cuad / n) >= 0.01:
        X = punto_i(np.random.random())
        n += 1

        promedio_prev = promedio
        promedio = prom_update(promedio, X, n)
        s_cuad = s_cuad_update(s_cuad, promedio_prev, promedio, n)

    print("Punto i")
    print(f"Integral = 2.0685 ~ {promedio:.4f}")
    print(f"Desviación estándar muestral = {math.sqrt(s_cuad / n):.4f} ")
    print(f"Cantidad de datos generados: ", n)

def sim_ii():
    # Se multiplica por 2 ya que la fun ii es simétrica y solo esta recibiendo
    # valores para la parte positiva
    promedio = punto_ii(np.random.random()) * 2
    s_cuad, n = 0, 1

    while n <= 100 or math.sqrt(s_cuad / n) >= 0.01:
        X = punto_ii(np.random.random()) * 2
        n += 1

        promedio_prev = promedio
        promedio = prom_update(promedio, X, n)
        s_cuad = s_cuad_update(s_cuad, promedio_prev, promedio, n)

    print("Punto ii")
    print(f"Integral = 0.8862 ~ {promedio:.4f}")
    print(f"Desviación estándar muestral = {math.sqrt(s_cuad / n):.4f} ")
    print(f"Cantidad de datos generados: ", n)

sim_i()
print()
sim_ii()
```
