### Ejercicio 5.
Sean $X_1 , \cdots , X_n$ variables aleatorias i.i.d., con media desconocida $\mu$. Para constantes $a \lt b$, se quiere estimar
$$
p = P(a \lt \frac{1}{n} \sum_{i=1}^{n} X_i - \mu \lt b)
$$

Estimar p si $n = 10$ y los valores de $x_i$ son $56, 101, 78, 67, 93, 87, 64, 72, 80, 69$. Tomar $a = −5, b = 5$.


---

#### Análisis
Se utilizara la técnica bootstrap. Como $N=10$ y sea $B = \{b_i : b_i \ \text{es una muestras bootstrap}\} \implies |B| = N^N = 10^{10} = 10.000.000.000$, lo cual es demasiado grande, por lo que se utilizara Monte Carlo para generar M = 10.000 muestras bootstrap y estimar p.

Notar que lo que deseamos estimar es lo siguiente:
$$ p = P(-5 \lt \hat{\theta} \lt 5) $$

Donde $\hat{\theta}$ es:
$$ \hat{\theta}(X_i) = \frac{1}{10} \sum_{i=1}^{10} X_i - \mu $$

Como se desconoce $\mu$, entonces utilizare:
$$\overline{\mu}_M = \frac{1}{M} \sum_{b_i \in B} \hat{\theta}_b(b_i), \quad \text{donde } \hat{\theta}_b(b_i) \text{ es la media muestral de la muestra bootstrap i}$$

```python
X_i = [56, 101, 78, 67, 93, 87, 64, 72, 80, 69]
M = 10_000

B = [] # Arreglo de muestras bootstrap
mu_M = 0
for _ in range(M):
    # Se genera una muestra aleatoria de muestra bootstrap
    b_i = np.random.choice(X_i, size=10)
    B.append(b_i)

    # Se va calculando el promedio (luego se lo divide por 10, ya que es una sumatoria)
    mu_M += sum(b_i)

# Calculo de la media muestral bootstrap
mu_M = mu_M / (10 * M)

data = np.array(X_i) - mu_M
casos_fav = sum((data > -5) & (data < 5))
casos_totales = 10

print("Estimación Mu = ", mu_M)
print("Estimación p = ", casos_fav / casos_totales)
```
