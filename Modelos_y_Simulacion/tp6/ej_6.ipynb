```python
import numpy as np


def s_cuadrado(n: int, media: float, x: list[int]):
    s = 0
    for i in range(n):
        s += (x[i] - media)**2

    return s / (n - 1)


def prom(n: int, x: list[int]):
    return sum(x) / n
```

### Ejercicio 6.
Sean $X_1 , \cdots , X_n$ variables aleatorias independientes e idénticamente distribuidas con varianza $\sigma^2$ desconocida. Se planea estimar $\sigma^2$ mediante la varianza muestral:
$$ S^2(n) = \frac{1}{n - 1} \sum_{i=1}^{n} (X_i - \overline{X}(n))^2 $$

a. Si $n = 2, X_1 = 1, X_2 = 3$, ¿cuál es la estimación “bootstrap” de $Var(S^2(n))$?

b. Si n = 15, los datos son:
$$ 5, 4, 9, 6, 21, 17, 11, 20, 7, 10, 21, 15, 13, 16, 8 $$
¿Cómo se calcula la estimación bootstrap en este caso? Dé un valor posible de la estimación.

---

#### Análisis
Sea $B = \{ b_i : b_i \ \text{es una muestra bootstrap} \}$.

##### Punto a
Como n = 2, entonces podemos calcular el valor directamente, ya que $|B| = n^n = 2^2 = 4$.

Para ello, se hará lo siguiente:

1. Se calculara $\hat{\theta}(b_i)$ (utilizando la $S^2(n)$, lo cual pide el ejercicio) de cada muestra bootstrap.
2. Se calculara la media muestral de cada $\hat{\theta}(b_i)$ ($\overline{\hat{\theta}}$).
3. Se calculara la varianza muestral con los datos de 1 y 2, el cual es la estimación “bootstrap” de $Var(S^2(n))$.

##### Punto b
Como n = 15. Tenemos que $n^n$ es un numero muy grande, por lo que se utilizara Monte Carlo con M = 10.000 para aproximarlo.

Para ello, se hará lo siguiente:

1. Se genera M $b_i$ para B.
2. Se calculara $\hat{\theta}(b_i)_M$ (utilizando la $S^2(n)$, lo cual pide el ejercicio) de cada muestra bootstrap.
3. Se calculara la media muestral de cada $\hat{\theta}_M(b_i)$.
4. Se calculara la varianza muestral con los datos de 1 y 2, el cual es la aprox a la estimación “bootstrap” de $Var(S^2(n))$.

```python
# Punto a
n = 2
x = [1, 3]

# Calculo todas las permutaciones de x
B = []
for i in range(n):
    for j in range(n):
        B.append([x[i], x[j]])

# Calculo todas las theta(b_i)
t_b_i = []
for i in range(n**n):
    mu_b_i = prom(n, B[i])
    s_cuad_b_i = s_cuadrado(n, mu_b_i, B[i])
    t_b_i.append(s_cuad_b_i)

# Calculo el promedio y luego la varianza de las theta(b_i)
mu = prom(n**n, t_b_i)
s_cuad = s_cuadrado(n**n, mu, t_b_i)

print("La estimación “bootstrap” de Var(S^2(n)) es ", s_cuad)
```

```python
#Punto b
n = 15
x = [5, 4, 9, 6, 21, 17, 11, 20, 7, 10, 21, 15, 13, 16, 8]
M = 10_000

# Genero muestras aleatorias de muestras bootstrap (b_i) y guardo los theta(b_i)
t_b = []
for _ in range(M):
    b_i = np.random.choice(x, size=n)

    mu_b_i = prom(n, b_i)
    s_cuad_b_i = s_cuadrado(n, mu_b_i, b_i)

    t_b.append(s_cuad_b_i)

# Calculo de la media muestral bootstrap
mu_M = prom(M, t_b)
s_cuad = s_cuadrado(M, mu_M, t_b)

print("Una aproximación a la estimación “bootstrap” de Var(S^2(n)) es ", s_cuad)
```
