```python
import numpy as np

def estimar_p_valor_simulaciones_continua(
    n: int,
    d_obs: float,
    N_SIM: int = 10_000
):
    # Se utiliza el estadístico de Kolmogorov-Smirnov, sin estimar parámetros
    
    p_valor = 0.0
    for _ in range(N_SIM):
        uniformes = np.random.uniform(0, 1, n)
        uniformes.sort()
        d_j = 0

        for j in range(n):
            u_j = uniformes[j]
            d_j = max(d_j, (j + 1) / n-u_j, u_j - j / n)

        if d_j >= d_obs:
            p_valor += 1

    p_valor = p_valor / N_SIM
    return p_valor
```

### Ejercicio 3.
Calcular una aproximación del *p−valor* de la hipótesis: “Los siguientes $10$ números son aleatorios”:
$$ 0.12, 0.18, 0.06, 0.33, 0.72, 0.83, 0.36, 0.27, 0.77, 0.74. $$

---

### Análisis
Tenemos que $H_0 = \text{"Los siguientes 10 números son aleatorios"} = \text{"Los datos provienen de una distribución uniforme entre 0 y 1"}$. Por lo cual $F = U(0, 1)$.

1. Primero tenemos que ordenar los datos.
2. Calculamos $D^+ = \frac{j}{n} - F(Y_j)$ y $D^- = F(Y_j) - \frac{j - 1}{n}$. Como los indices ($i$) comienzan desde 0, utilizare $j = i + 1$
3. Calculamos $D = max(D^+, D^-)$.
4. Asumiendo un $\alpha = 0.05$, si $\text{p-valor} \le \alpha$ entonces rechazo $H_0$

```python
x = [0.12, 0.18, 0.06, 0.33, 0.72, 0.83, 0.36, 0.27, 0.77, 0.74]
x.sort()

n = len(x)

def F(x: float):
    # La acumulada de la uniforme continua para valores entre 0 y 1 es x
    return x

D_p = [((i + 1)/n - F(x[i])) for i in range(n)]
D_m = [(F(x[i]) - i/n) for i in range(n)]

d_obs = max(max(D_p), max(D_m))
p_valor = estimar_p_valor_simulaciones_continua(n, d_obs)

print("d_obs: ", d_obs)
print("p-valor: ", p_valor)
```

Como el p-valor es bastante mayor a $\alpha$, entonces no existe evidencia suficiente para rechazar $H_0$.
