# Ejercicio 2

Para verificar que cierto dado no estaba truncado, se registraron $1000$ lanzamientos, resultando que el número de veces que el dado arrojó el valor $i$ $(i=1,2,3,4,5,6)$ fue, respectivamente, $158, 172, 164, 181, 160, 165$.

Aproximar el *p-valor* de la prueba: "el dado es honesto".

**a)** Utilizando la prueba de Pearson con aproximación chi-cuadrada.

**b)** Realizando una simulación.

**Resolución:**

Queremos testear la hipótesis que estos datos provienen de una distribución uniforme discreta $X\sim U\{1,6\}$. Nuestra hipótesis nula:
$$
H_0: \text{El dado es honesto}.
$$
es decir, que $P(X=i)=\frac{1}{6}$ para cada $i\in \{1,2,3,4,5,6 \}$


Tenemos un problema con **datos discretos** y **parámetros conocidos**.



**Inciso a)**

Usamos la prueba chi-cuadrada Pearson con el estadístico:
$$
T = \sum_{i=1}^{k=6}\frac{(N_i - n p_i)^2}{n p_i}
$$

Para cada $i\in \{1,2,3,4,5,6 \}$:


*   $N_i$ es la frecuencia observada en los datos correspondientes a la clase $i$.
*   $p_i$ es la probabilidad de que la variable aleatoria tome el valor $i$



El *p-valor* $p$ viene dado por:
$$
p = P(T \geq t_{obs}) = P(χ_{k-1} \geq t_{obs}) =  P(χ_5 \geq t_{obs})
$$

Nuestro $k$ es igual a $6$ porque la variable $X$ puede tomar $6$ valores distintos.

```python
from scipy.stats import chi2
import numpy as np
```

```python
def estadisticoT(frecuencias, probabilidades):
    t = 0
    K = len(frecuencias)
    n = sum(frecuencias)
    for k in range(K):
        t = t + (frecuencias[k] - n * probabilidades[k])**2 / (n * probabilidades[k])
    return t
```

```python
probabilidades = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]
frecuencias = [158, 172, 164, 181, 160, 165]

print('Cantidad de datos: ', sum(frecuencias))
```

```python
t0 = estadisticoT(frecuencias, probabilidades)
print('p-valor: ', 1 - chi2.cdf(t0, df=5))     # df=5 porque tengo k-1 grados de libertad y k=6
```

**Inciso b)**

Realizamos una simulación para estimar el $p$-valor.

Para esto necesitamos:

1. El $t_{obs}$ de nuestra muestra de datos.

2. Generar $Nsim$ muestras con la misma cantidad de datos de la muestra originar y calcular el valor del estadístico $T$ en cada una de esas muestras.

3. El $p$-valor se aproxima como la proporción de valores que superan el valor $t_{obs}$.

```python
def generar_numero_discreto(p, x):
    '''
    x: lista de valores posibles de la variable aleatoria
    p: lista de probabilidades asociadas a cada valor de x
    '''
    u = np.random.random()
    i = 0
    F = p
    while u >= F:
        i = i+1
        F = F + 1/6
    return x[i]

def frecuencias_generadas(n):
    freq = [0,0,0,0,0,0]
    for k in range(n):
        x = generar_numero_discreto(1/6, [1,2,3,4,5,6])
        freq[x-1] = freq[x-1] + 1
    return freq

def ejercicio_2b(frecuencias, probabilidades, NSim):
    pvalor = 0
    n = sum(frecuencias)                                # cantidad de datos de mi muestra
    t_obs = estadisticoT(frecuencias, probabilidades)
    for k in range(NSim):
        freq = frecuencias_generadas(n)                 # frecuencias de las clases en mi muestra generada con distribución U{1,6}
        t = estadisticoT(freq, probabilidades)          # calculo el estadístico para la muestra generada
        if t_obs <= t:
            pvalor = pvalor + 1
    return pvalor/NSim
```

```python
probabilidades = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]
frecuencias = [158, 172, 164, 181, 160, 165]
Nsim = 10000
t = ejercicio_2b(frecuencias, probabilidades, Nsim)

print('Ejercicio b:')
print('p-valor estimado: ', t)
```

# Ejercicio 5

Calcular una aproximación del $p-$valor de la prueba de que los siguientes datos corresponden a una distribución binomial con parámetros $(n=8, p)$, donde $p$ no se conoce:

$$ 6, 7, 3, 4, 7, 3, 7, 2, 6, 3, 7, 8, 2, 1, 3, 5, 8, 7.$$

$H_0$: los datos observados provienen de una distribución binomial con parámetros $n=8$ y $p$ desconocido.

Tenemos un problema con **datos discretos y un parámetro desconocido**


**Estimación del parámetro**

Sabemos que si $X \sim B(n,p)$, $E[X] = np$. En nuestro caso, $E[X]=8p$, y podemos estimar $p$ mediante la media muestral:

$$
\hat p=\frac{\bar X}{8}.
$$


**Test $\chi²$ de Pearson**


Una vez estimado $\hat p$, tenemos  una distribución $\hat F = Bin(n, \hat p)$, y las probabilidades teóricas son

$$
\hat p_k = P_{\hat F}(X=k) = \binom{8}{k} \ \hat p^k \ (1-\hat p)^{8-k},
\qquad k=0,\ldots,8.
$$


Sea

$$
N_k=\#\{X_i=k\},
\qquad k=0,\ldots,8,
$$
la frecuencia observada de cada categoría.

Luego, el estadístico de Pearson es

$$
T= \sum_{k=0}^{8} \frac{(N_k-n \ \hat p_k)^2}{n \ \hat p_k}.
$$

Si $m$ es el numero de parámetros que se utilizan para el cálculo de $p_i$ y que deben ser estimados, vimos en el teórico que, para $n$ suficientemente grande, el estadístico $T$ tiene aproximadamente una distribucion $\chi ²$ con $k-1-m$ grados de libertad.

En nuestro caso, como la variable puede tomar $9$ posibles valores y se estima un solo parámetro → $k=9, \, m=1$. Luego:

$$
T\sim \chi^2_{7}.
$$

Y el p-valor puede estimarse como:

$$
p\text{-valor} = P(\chi^2_7\ge t_{obs}).
$$


```python
import numpy as np
from scipy.stats import chi2, binom

datos = [6,7,3,4,7,3,7,2,6,3,7,8,2,1,3,5,8,7]
n = len(datos)

p_est = np.mean(datos)/8    # estimador de p
N = np.array([datos.count(k) for k in range(9)])  # construye las frecs observadas
# tambièn puede ser: np.bincount(datos, minlength=9)

prob = np.array([binom.pmf(k,8,p_est) for k in range(9)])   # calcula las probs teòricas de la binomial(n, p^)

E = n*prob  # frecuecias esperadas

T = np.sum((N-E)**2/E)  # estadístico (N-np)^2
pvalor = chi2.sf(T, df=7)   # P(χ^2_7 ​≥ T)

print("p estimado =", p_est)
print("T =", T)
print("p-valor =", pvalor)
```

**Aproximación mediante simulación**

También podemos aproximar el p-valor mediante simulación.

Bajo la hipótesis nula, los datos provienen de una distribución $X \sim Bin(8,\hat p)$, donde $\hat p$ es el valor estimado a partir de la muestra observada.

Se realizan $N_{sim}$ simulaciones independientes. En cada simulación:

1. Se genera una muestra de tamaño $n=18$ de una distribución $Bin(8,\hat p)$.
2. Se reestima el parámetro $p$, ya que en el problema original dicho parámetro es desconocido y fue estimado a partir de la muestra.
$$ \hat p^{(j)} = \frac{\bar X^{(j)}}{8}. $$

3. Se calculan las probabilidades teóricas

$$
p_k^{(j)} = \binom{8}{k} \left(\hat p^{(j)}\right)^k \left(1-\hat p^{(j)}\right)^{8-k},
\qquad k=0,\ldots,8.
$$

4. Se obtienen las frecuencias esperadas

$$
E_k^{(j)} = n\,p_k^{(j)}.
$$

5. Se calcula el estadístico de Pearson

$$
T_{(j)} = \sum_{k=0}^{8} \frac{\left( N_k^{(j)}-n\,p_k^{(j)} \right)^2}{n\,p_k^{(j)}}.
$$

Si $t_{obs}$ es el estadístico obtenido con la muestra original, el p-valor se aproxima mediante

$$
p\text{-valor} \approx \frac{\#\{j:t_{(j)}\ge t_{obs}\}}{N_{sim}}.
$$

```python
import numpy as np
from scipy.stats import binom

Nsim = 10000
count = 0

for _ in range(Nsim):

    muestra = np.random.binomial(n=8, p=p_est, size=n) # muestra simulada bajo H0
    p_sim = np.mean(muestra)/8  # reestimar p

    # frecuencias observadas
    # N_sim = np.bincount(muestra, minlength=9)
    N_sim = np.array([np.sum(muestra == k) for k in range(9)])

    # frecuencias esperadas
    prob_sim = np.array([binom.pmf(k,8,p_sim) for k in range(9)])

    E_sim = n*prob_sim  # E = np
    T_sim = np.sum((N_sim-E_sim)**2/E_sim)   # estadìstico

    if T_sim >= T:
        count += 1

pvalor_sim = count/Nsim

print("p-valor simulado =", pvalor_sim)
```
