# Ejercicio 2

Se propone el siguiente juego en el cual todas las variables aleatorias que se generan son independientes e idénticamente distribuidas  $\mathcal{U}(0,1)$: Se simula la variable aleatoria $U$. Si $U<\displaystyle\frac{1}{2}$, se suman dos nuevos números aleatorios $W_1+W_2$. Pero si $U\geq\displaystyle\frac{1}{2}$, se suman tres números aleatorios. El resultado de la suma, en cualquiera de los casos, es una variable aleatoria $X$. Se gana en el juego si $X\geq1$.

**a)** ¿Cuál es la probabilidad de ganar?.

**b)** Implementar un algoritmo en computadora que estime la probabilidad de ganar, esto es, la fracción de veces que se gana en $n$ realizaciones del juego. Completar la siguiente tabla:

$$
\begin{array}{|c|c|c|c|c|c|}
\hline
n & 100 & 1000 & 10000 & 100000 & 1000000 \\ \hline
P[X \geq 1] & & & & & \\ \hline
\end{array}
$$

### Probabilidad de ganar

Por probabilidad total:

$$
P(\text{ganar}) = P(\text{ganar}\mid U < 0.5)P(U < 0.5) + P(\text{ganar}\mid U\ge 0.5)P(U\ge 0.5)
$$

Como $P(U < 0.5)=P(U\ge0.5)=0.5$:

$$
P(\text{ganar}) = 0.5\,P(W_1+W_2 \ge 1) + 0.5\,P(W_1+W_2+W_3 \ge 1)
$$

---

**Caso 1:** dos variables

$$
P(W_1+W_2 \ge 1)
= \int_0^1 \int_{1-w_2}^1 dw_1\,dw_2
= \frac{1}{2}
$$

---

**Caso 2:** tres variables

Aplicamos probabilidad total condicionando en $W_1+W_2$:

$$
P(W_1+W_2+W_3 \ge 1)
$$

$$
= P(W_1+W_2+W_3 \ge 1 \mid W_1+W_2 \ge 1)P(W_1+W_2 \ge 1)
+ P(W_1+W_2+W_3 \ge 1\mid W_1+W_2 < 1)P(W_1+W_2 < 1)
$$

---

- Primer término

Si $W_1+W_2 \ge 1$, entonces:

$$
W_1+W_2+W_3 \ge 1
$$

siempre, luego:

$$
P(W_1+W_2+W_3 \ge 1 \mid W_1+W_2 \ge 1) = 1
\quad,\quad P(W_1+W_2 \ge 1)=\frac{1}{2}
$$

---
- Segundo término

$$
P(W_1+W_2+W_3 \ge 1,\; W_1+W_2 < 1)
$$

corresponde al volumen:

$$
\int_0^1 \int_0^{1-w_1} \int_{1-w_1-w_2}^1 dw_3\,dw_2\,dw_1
= \frac{1}{3}
$$

Entonces:

$$
P(W_1+W_2+W_3 \ge 1 \mid W_1+W_2 < 1)
= \frac{1/3}{1/2}
= \frac{2}{3}
$$

---

Resultado del caso 2

$$
P(W_1+W_2+W_3 \ge 1)
= 1 \cdot \frac{1}{2} + \frac{2}{3}\cdot \frac{1}{2}
= \frac{1}{2} + \frac{1}{3}
= \frac{5}{6}
$$

---
Resultado final

$$
P(\text{ganar})
= 0.5 \cdot \frac{1}{2} + 0.5 \cdot \frac{5}{6}
= \frac{1}{4} + \frac{5}{12}
= \frac{2}{3}
$$

---

$$
\boxed{P(\text{ganar}) = \frac{2}{3}}
$$

```python
import random
import numpy as np

# un experimento
def experimento():
    u = random.random()

    if u < 0.5:
        x = random.random() + random.random()
    else:
        x = random.random() + random.random() + random.random()

    return int(x >= 1)

# estimación
def estimar_probabilidad(Nsim):
    return np.mean([experimento() for _ in range(Nsim)])

def estimar_probabilidad2(Nsim):
    exitos = 0

    for _ in range(Nsim):
        resultado = experimento()

        if resultado == 1:
            exitos += 1

    probabilidad = exitos / Nsim

    return probabilidad


# tabla
ns = [100, 1000, 10000, 100000, 1000000]

for n in ns:
    print(f"n = {n}: {estimar_probabilidad2(n)}")

print("\nValor teórico:", 2/3)
```
