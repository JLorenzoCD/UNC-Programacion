### Ejercicio 1.
Desarrolle un método para generar una variable aleatoria cuya densidad de probabilidad es:

a)
$$
f(x) = \left\{
\begin{array}{cl}

\frac{x - 2}{2} & : \ 2 \lt x \le 3 \\
\\
\frac{2 - x/3 }{2} & : \ 3 \lt x \le 6 \\
\\
0 & : \ \text{en otro caso}

\end{array}
\right.
$$

```python
def F_inv_a(u: int):
    if u < 0.25:
        return math.sqrt(u) + 2
    else:
        return 6 - math.sqrt(12 * (1 - u))

def varX_a():
    U = np.random.random()
    return F_inv_a(U)


print([varX_a() for _ in range(10)])
```

b)
$$
f(x) = \left\{
\begin{array}{cl}

\frac{6(x+3)}{35} & : \ 0 \lt x \le 1 \\

\\
\frac{6x^{2}}{35} & : \ 1 \lt x \le 2 \\

\\
0 & : \ \text{en otro caso}

\end{array}
\right.
$$

```python
def F_inv_b(u: int):
    if u < 0.6:
        return math.sqrt(9 + u * (35/3)) - 3
    else:
        return math.cbrt((35 * u - 19) / 2)

def varX_b():
    U = np.random.random()
    return F_inv_b(U)


print([varX_b() for _ in range(10)])
```

c)
$$
f(x) = \left\{
\begin{array}{cl}

\frac{\exp(4x)}{4} & : \ -\infty  \lt x \le 0 \\ \\
\frac{1}{4} & : \ 0 \lt x \le \frac{15}{4} \\\\
0 & : \ \text{en otro caso}

\end{array}
\right.

$$

```python
def F_inv_c(u: int):
    if u < 0.0625: # 1/16
        return math.log(16 * (1 - u)) / 4
    else:
        return 4 * u - 0.25

def varX_c():
    U = np.random.random()
    return F_inv_c(U)


print([varX_c() for _ in range(10)])
```
