```python
import math
import time
from typing import Callable
import numpy as np
```

### Ejercicio 8.
a) Sean U y V dos variables aleatorias uniformes en (0, 1) e independientes. Pruebe que la variable
$X = U + V$ tiene una densidad triangular:

$$
f(x) = \left\{
\begin{array}{cl}

x & : \ 0 \le x \lt 1 \\\\
2 - x & : \ 1 \le x \lt 2 \\\\
0 & : \ \text{en otro caso}

\end{array}
\right.
$$
b. Desarrolle tres algoritmos que simulen la variable X:
1. Usando la propiedad que X es suma de dos uniformes.
2. Aplicando transformada inversa.
3. Con el método de rechazo.

c. Compare la eficiencia de los tres algoritmos. Para cada caso, estimar el valor esperado promediando 10.000 valores simulados, ¿Para qué valor $x_0$ se cumple que $P(X \gt x_0) = 0.125$?

d. Compare la proporción de veces que el algoritmo devuelve un número mayor que $x_0$ con esta probabilidad.

---

```python
def var_X_SUM():
    U1, U2 = np.random.random(), np.random.random()
    return U1 + U2

def var_X_TI():
    U1, U2 = np.random.random(), np.random.random()

    if U1 < 0.5:
        return math.sqrt(2 * U2)
    else:
        return 2 - math.sqrt(2 * U2)

def var_X_AyR():
    def f(x: int):
        if x < 0:
            return 0
        elif x < 1:
            return x
        elif x < 2:
            return 2 - x
        else:
            return 0

    while True:
        Y = np.random.uniform(0, 2)
        U = np.random.random()

        if U < f(Y): # C = 2
            return Y


def simulaciones(generador: Callable[[], float], N_SIM: int = 10_000):
    start = time.time()
    samples = np.array([generador() for _ in range(N_SIM)])
    t = time.time() - start

    e = 1
    est = samples.mean()

    p_lq_3_2 = sum(samples <= 1.5) / N_SIM
    p = 0.875

    print(f"Generador: {generador.__name__:>9} - time: {t:.6f}")
    print(f"E[X] = {e:.6f} ~ {est:.6f}")
    print(f"P(X <= 3/2) = {p:.6f} ~ {p_lq_3_2:.6f}")

simulaciones(var_X_SUM)
print()
simulaciones(var_X_TI)
print()
simulaciones(var_X_AyR)
```
