### Ejercicio 6.
Utilice el método del rechazo y los resultados del ejercicio anterior para desarrollar otros dos métodos, además del método de la transformada inversa, para generar una variable aleatoria con distribución de probabilidad

$$
F(x) = x^n , \quad 0 \le x \le 1 .
$$

Analice la eficiencia de los tres métodos para generar la variable a partir de F.

---

#### Resolución
- Método de la Transformada inversa:
$$F^{-1}(u) = u^{1/n}$$
- Utilizando F_M (ej 5):
Se genera n uniformes y se elige el máximo
- Método de aceptación y rechazo:
$$F'(x) = f(x) = n.x^{n-1}, \quad 0 \le x \le 1$$
Entonces tomo $Y \sim U(0, 1)$ con $g(x) = 1$. Lo cual me da que:
$$
\frac{f(x)}{g(x)} \le c = \frac{n.x^{n-1}}{1} \le c
$$
Como el max valor de x = 1
$$\frac{n.1^{n-1}}{1} \le c = n \le c \implies c = n$$
Lo cual nos da la siguiente cota:
$$U \lt \frac{f(Y)}{g(Y).c} = U \lt \frac{n.Y^{n-1}}{1.n} = U \lt Y^{n-1}$$

```python
def var_X_TI(n: int):
    U = np.random.random()

    return U**(1/n)

def var_X_M(n: int):
    M = 0
    for _ in range(n):
        U = np.random.random()
        M = max(M, U)

    return M

def var_x_AyR(n: int):
    while True:
        U, Y = np.random.random(), np.random.random()

        if U < Y**(n - 1):
            return Y

def t_gen(n: int, generador: Callable[[int], float]):
    start = time.time()
    muestras = [generador(n) for _ in range(n)]
    end = time.time()

    t = end - start
    # print(muestras)
    print(f"N = {n}, tiempo = {t:.6f}, generador = {generador.__name__}")

t_gen(1_000, var_X_TI)
t_gen(1_000, var_X_M)
t_gen(1_000, var_x_AyR)
```
