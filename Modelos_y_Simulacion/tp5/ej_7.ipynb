```python
import math
import time
from typing import Callable
import numpy as np
```

### Ejercicio 7.
a. Desarrolle dos métodos para generar una variable aleatoria X con densidad de probabilidad:

$$
f(x) = \left\{
\begin{array}{cl}

\frac{1}{x} & : \ 1 \le x \le e \\
\\
0 & : \ \text{en otro caso}

\end{array}
\right.
$$
1. Aplicando Transformada inversa.

2. Aplicando el método de aceptación y rechazo con una variable uniforme.

b. Compare la eficiencia de ambos métodos realizando 10.000 simulaciones y comparando el promedio de los valores obtenidos. Compruebe que se obtiene un valor aproximado del valor esperado de X.

c. Estime la probabilidad P(X ≤ 2) y compárela con el valor real.

---

```python
def var_X_TI():
    U = np.random.random()
    return math.exp(U)

def var_X_AyR():
    while True:
        Y = np.random.uniform(1, math.e)
        U = np.random.random()

        if U < 1 / (Y * 1.17183):
            return Y


def simulaciones(generador: Callable[[], float], N_SIM: int = 10_000):
    start = time.time()
    samples = np.array([var_X_TI() for _ in range(N_SIM)])
    t = time.time() - start

    e = math.e - 1
    est = samples.mean()

    p_lq_2 = sum(samples <= 2) / N_SIM
    p = math.log(2)

    print(f"Generador: {generador.__name__:>9} - time: {t:.6f}")
    print(f"E[X] = {e:.6f} ~ {est:.6f}")
    print(f"P(X <= 2) = {p:.6f} ~ {p_lq_2:.6f}")

simulaciones(var_X_TI)
print()
simulaciones(var_X_AyR)
```
