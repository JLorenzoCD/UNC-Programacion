b) Genere datos usando tres exponenciales independientes con media 3, 5 y 7 respectivamente y p = (0.5, 0.3, 0.2). Calcule la esperanza exacta de la mezcla y estime con 10.000 repeticiones. Tenga cuidado con la parametrización que este usando!!

```python
def punto_b():
    U1 = np.random.random()
    media = 0

    if U1 < 0.5:
        media = 3
    elif U1 < 0.8:
        media = 5
    else:
        media = 7

    U2 = np.random.random()
    return - math.log(1 - U2) * media  # - Ln(U) / lambda = - Ln(U) * media

N_SIM = 10_000
datos = np.array([punto_b() for _ in range(N_SIM)])
esperanza = 4.4  # 3 * 0.5 + 5 * 0.3 + 7 * 0.2
est_esperanza = datos.mean()
print(f"N° simulaciones: {N_SIM}")
print(f"E[X] = {esperanza} ~ {est_esperanza:.4f}")
```
