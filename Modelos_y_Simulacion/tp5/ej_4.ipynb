### Ejercicio 4.
Desarrolle un método para generar la variable aleatoria con función de distribución

$$
F(x) = \int _{\theta }^{\infty } x^y e^{-y} dy, \qquad 0 \leq x \leq 1
$$

Piense en el método de composición del ejercicio anterior. En particular, sea F la función de distribución de X y suponga que la distribución condicional de X dado Y = y es

$$
P(X \leq x \ |\ Y = y) = x^y , \qquad 0 \leq x \leq 1
$$

#### Resolución
Notar que Y seria una v.a. con $Y \sim  \huge \varepsilon \normalsize(\lambda = 1)$

$$
F(x) = \int _{\theta }^{\infty } F_{X|Y}(x|y) \ f_Y(y) dy, \qquad 0 \leq x \leq 1
$$

Entonces, tendríamos $f_Y(y)$ seria equivalente al $p_i$ y que $F_{X|Y}(x|y)$ seria equivalente a $F_i(x)$ en el Ejercicio 3.

$$
F_{X|Y}^{-1}(u) = u^{1/y}, \qquad 0 \leq x \leq 1
$$

```python
def varX():
    U1, U2 = np.random.random(), np.random.random()
    Y = - math.log(1 - U1) # Se genera Y, la cual es una v.a. exponencial con lambda = 1

    return U2**(1/Y) # inversa de F_{X|Y}
```
