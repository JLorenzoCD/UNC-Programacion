```python
import math
import time
from typing import Callable
import numpy as np
```

### Ejercicio 9.
a. generación de variables normales según el ejemplo 5 f del libro Simulacion de S. M. Ross,

b. el método polar,

c. el método de razón entre uniformes.

Pruebe los códigos calculando la media muestral y varianza muestral de 10.000 valores generados con los
tres métodos.

---

```python
def punto_a_MAyR(mu: float = 0.0, sigma: float = 1.0):
    while True:
        Y1 = -math.log(1 - np.random.random())
        Y2 = -math.log(1 - np.random.random())

        if Y2 >= (Y1 - 1) ** 2 / 2:
            if np.random.random() < 0.5:
                return Y1 * sigma + mu

            return -Y1 * sigma + mu


def punto_b_MP(mu: float = 0.0, sigma: float = 1.0):
    Rcuadrado = -2 * math.log(1 - np.random.random())
    Theta = 2 * math.pi * np.random.random()

    X = math.sqrt(Rcuadrado) * math.cos(Theta)
    Y = math.sqrt(Rcuadrado) * math.sin(Theta)

    return (X * sigma + mu, Y * sigma + mu)


NV_MAGICCONST = 4 * math.exp(-0.5) / math.sqrt(2.0)
def punto_c_MReU(mu: float = 0.0, sigma: float = 1.0):
    while True:
        u1 = np.random.random()
        u2 = 1 - np.random.random()

        z = NV_MAGICCONST * (u1 - 0.5) / u2
        zz = z * z / 4
        if zz <= -math.log(u2):
            break

    return mu + z * sigma

def print_sim(N_SIM: int, time: float, func_name: str, e: float, v: float):
    print(
        f"N° sim = {N_SIM:>7} - Tiempo = {time:.4f} - Fun = {func_name:>12} - E[X] = {e:.4f} - Var(X) = {v:.4f}")


def simulaciones(generador: Callable[[float, float], float], N_SIM: int = 10_000):
    start = time.time()
    muestras = np.array([generador() for _ in range(N_SIM)])
    t = time.time() - start

    print_sim(N_SIM, t, generador.__name__, muestras.mean(), muestras.var())


def simulaciones_doble(generador: Callable[[float, float], tuple[float, float]], N_SIM: int = 5_000):
    start = time.time()
    muestras = []
    for _ in range(N_SIM):
        X, Y = generador()
        muestras.append(X)
        muestras.append(Y)
    t = time.time() - start

    muestras = np.array(muestras)
    print_sim(N_SIM, t, generador.__name__, muestras.mean(), muestras.var())


simulaciones(punto_a_MAyR)
simulaciones_doble(punto_b_MP)
simulaciones(punto_c_MReU)
```
