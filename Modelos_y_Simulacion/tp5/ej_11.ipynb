```python
import math
import time
from typing import Callable
import numpy as np
```

### Ejercicio 11.
Sea X una variable aleatoria con distribución de Cauchy de parámetro λ.

a. Calcule la función de distribución acumulada $F_X$.

b. Simule X aplicando el método de transformada inversa.

c. Indique si es posible generar X por el método de aceptación y rechazo, rechazando con una variable aleatoria normal.

d. Realice 10000 simulaciones y calcular la proporción de veces que el resultado cae en el intervalo (−λ, λ), para λ = 1, λ = 2.5 y λ = 0.3. Compare con la probabilidad teórica.

e. Compare la eficiencia de este algoritmo con el método de razón entre uniformes.

---

```python
def CAUCHY_TI(lamda: float = 1.0):
    U = np.random.random() # U no puede ser 0 ni 1
    while U == 0:
        U = np.random.random()

    return math.tan((U - 1/2) * math.pi) * lamda

def CAUCHY_MReU(lamda: float = 1.0):
    while True:
        # Se genera un semicírculo derecho centrado en (0,0) (cuadrante 1 y 4)

        # Por def, U > 0
        U = np.random.random()

        # -1 < V < 1
        # Se podría generar V entre (0, 1), pero luego se tendria que generar U2
        # para poner signo + o - con una chance de U2 < 0.50
        V = np.random.random() * 2 - 1

        if U**2 + V**2 < 1:
            return (V / U) * lamda

def simulaciones(generador: Callable[[float], float] ,lamda: float = 1.0, N_SIM: int = 10_000):
    start = time.time()
    muestras = np.array([generador(lamda) for _ in range(N_SIM)])
    t = time.time() - start

    r = (muestras > -lamda) & (muestras < lamda)

    print(
        f"λ = {lamda:<3} - P(-λ < X < λ) = 0.5000 ~ {r.mean():.4f} - Tiempo: {t:.6f}")


print(f"Método: {CAUCHY_TI.__name__}")
simulaciones(CAUCHY_TI, 1.0)
simulaciones(CAUCHY_TI, 2.5)
simulaciones(CAUCHY_TI, 0.3)

print()
print(f"Método: {CAUCHY_MReU.__name__}")
simulaciones(CAUCHY_MReU, 1.0)
simulaciones(CAUCHY_MReU, 2.5)
simulaciones(CAUCHY_MReU, 0.3)
```
