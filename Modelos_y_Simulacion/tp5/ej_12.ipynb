### Ejercicio 12.
Escriba un programa que calcule el número de eventos y sus tiempos de arribo en las primeras T unidades de tiempo de un proceso de Poisson homogéneo con parámetro λ.

---

```python
def PPosson_homogeneo(lamda: float, T: float) -> tuple[int, list[float]]:
    eventos = []
    t = -math.log(1 - np.random.random()) / lamda

    while t < T:
        eventos.append(t)
        t -= math.log(1 - np.random.random()) / lamda

    return len(eventos), eventos
```
