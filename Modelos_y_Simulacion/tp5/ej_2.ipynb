a) Desarrolle métodos para generar las siguientes variables aleatorias

##### i) Distribución Pareto

$$ f(x) = ax^{-(a + 1)}, \qquad 1 \le x \lt \infty ,\quad a \gt 0 $$

```python
def gen_pareto(A: int):
    U = np.random.random()

    return U**(-1/A)


print(gen_pareto(1))
```

##### ii)  Distribución Erlang
$$
f(x) = \frac{x^{k - 1} \exp(-x/\mu)}{(k - 1)! \mu^{k}}, \qquad
0 \le x \lt \infty ,\quad
\mu \gt 0, \quad
k \in \mathbb{Z}
$$

```python
def gen_erlang(MU: float, K: int):
    prod = 1
    for _ in range(K):
        U = np.random.random()
        prod *= (1 - U)

    return - MU * math.log(prod)

print(gen_erlang(2, 8))
```

##### iii)  Distribución Weibull
$$
f(x) =
    \frac{\beta}{\lambda}
    \left(
        \frac{x}{\lambda}
    \right)^{\beta - 1}
    \exp(- (x/\lambda)^{\beta}),
    \qquad
0 \le x,\quad
\lambda \gt 0, \quad
\beta \gt 0
$$

```python
def gen_weibull(_BETA: float, _LAMBDA: float):
    U = np.random.random()
    l = math.log((_BETA - U*(_BETA - 1)) / _BETA)
    return _LAMBDA * ((- l)**(1 / _BETA))

print(gen_weibull(0.1, 0.5))
```
