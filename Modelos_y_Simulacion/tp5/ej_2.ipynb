### Ejercicio 2.
a) Desarrolle métodos para generar las siguientes variables aleatorias

##### i) Distribución Pareto

$$ f(x) = ax^{-(a + 1)}, \qquad 1 \le x \lt \infty ,\quad a \gt 0 $$

```python
def gen_pareto(A: int):
    U = np.random.random()

    return U**(-1/A)
```

##### ii)  Distribución Erlang
$$
f(x) = \frac{x^{k - 1} \exp(-x/\mu)}{(k - 1)! \mu^{k}}, \qquad
0 \le x \lt \infty ,\quad
\mu \gt 0, \quad
k \in \mathbb{Z}
$$

```python
def gen_erlang(MU: float, K: int):
    prod = 1
    for _ in range(K):
        U = np.random.random()
        prod *= (1 - U)

    return - MU * math.log(prod)
```

##### iii)  Distribución Weibull
$$
f(x) =
    \frac{\beta}{\lambda}
    \left(
        \frac{x}{\lambda}
    \right)^{\beta - 1}
    \exp(- (x/\lambda)^{\beta}),
    \qquad
0 \le x,\quad
\lambda \gt 0, \quad
\beta \gt 0
$$

```python
def gen_weibull(_LAMBDA: float, _BETA: float):
    U = np.random.random()
    return _LAMBDA * ((- math.log(1 - U))**(1 / _BETA))
```

b) Estime la media de cada variable con 10.000 repeticiones, usando los parámetros a = 2, µ =
2, k = 2, λ = 1, β = 2. Busque en la web los valores de las esperanzas para cada variable
con estos parámetros (cuidado con las parametrizaciones) y compare los valores obtenidos.

```python
def estimaciones(N_SIM: int = 10_000, _A: int = 2, _MU: float = 2, _K: int = 2, _LAMBDA: float = 1, _BETA: float = 2):
    muestras_i = np.array([gen_pareto(_A) for _ in range(N_SIM)])
    muestras_ii = np.array([gen_erlang(_MU, _K) for _ in range(N_SIM)])
    muestras_iii = np.array([gen_weibull(_LAMBDA, _BETA) for _ in range(N_SIM)])

    esp_pareto = _A / (_A - 1)
    esp_erlang = _K * _MU
    esp_weibull = _LAMBDA * 0.5 * math.sqrt(math.pi)
    print(
        f"X ~ Pareto(a=2) \t => E[X] = {esp_pareto} \t\t ~ {muestras_i.mean():.6f}")
    print(
        f"X ~ Erlang(µ=2, k=2) \t => E[X] = {esp_erlang} \t\t ~ {muestras_ii.mean():.6f}")
    print(
        f"X ~ Weibull(λ=1, β=2) \t => E[X] = {esp_weibull:.6f} \t ~ {muestras_iii.mean():.6f}")

estimaciones()
```
